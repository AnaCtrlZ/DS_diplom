```python
import pandas as pd
import os
from tqdm.notebook import tqdm
```

```python
def table(input_string, name_columns, name_table):
    # Разделяем строку на отдельные значения
    values = input_string.split()

    # Преобразуем список значений в список списков, где каждый внутренний список содержит три значения
    rows = [values[i:i+3] for i in range(0, len(values), len(name_columns))]

    # Создаем DataFrame из списка строк
    df = pd.DataFrame(rows, columns = list(name_columns))

    # Присваиваем имя таблице
    df.name = name_table

    # Выводим название таблицы и сам DataFrame
    print(f"Название таблицы: {df.name}")
    print(df)

    filename = name_table + ".md"
    counter = 1

    # Проверяем, существует ли файл с таким именем
    while os.path.exists(filename):
        filename = f"{name_table}_{counter}.md"
        counter += 1

    # Сохраняем таблицу в формате Markdown
    with open(filename, "w", encoding="utf-8") as file:
        file.write(f"# {df.name}\n\n")  # Заголовок с именем таблицы
        file.write(df.to_markdown(index=False))  # Сама таблица

    print(f"Таблица сохранена в файл: {filename}")
```

```python
name_columns = "Число жил, номинальное сечение (мм2), исполнение, напряжение (кВ)", "Расчетный диаметр кабеля, мм", "Расчетная масса кабеля, кг/км"
name_table = "ПБПнг(А)-FRHF FE180 E90"
```

```python
def rename(folder, number):
    for md_file in tqdm(os.listdir(folder)):
       md_path = os.path.join(folder, md_file)
       os.rename(md_path, os.path.join(folder, f"page_{number}"))
       number = number + 1
```
